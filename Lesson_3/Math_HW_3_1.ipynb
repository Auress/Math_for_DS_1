```python
print('Для построения окружности введите координаты центра и радиус через запятую (x0, y0, R):')
x0, y0, R = [float(x) for x in input().split(',')]

#x0 = -3
#y0 = -2
x = []
y1 = []
y2 = []
y3 = []
y4 = []
#R = 5

fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(8, 8)
plt.axis([-R*2, R*2, -R*2, R*2])
plt.grid()

for i in range(-1001, 1001):
    x_ = (i / 1000) * R + x0
    x.append(x_)
    y1.append((math.sqrt(abs(R**2 - (x_-x0)**2))) + y0)
    y2.append(-((math.sqrt(abs(R**2 - (x_-x0)**2))) - y0))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
print('Для построения элипса введите координаты центра, размеры a и b (x0, y0, a, b):')
x0, y0, a, b = [float(x) for x in input().split(',')]

#x0 = -2
#y0 = -3
x = []
y1 = []
y2 = []
#a = 5
#b = 2

fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(8, 8)
plt.axis([-(max(a,b)*2+1), (max(a,b)*2+1), -((max(a,b)*2+1)), (max(a,b)*2+1)])
plt.grid()

for i in range(-1001, 1001):
    x_ = (i / 1000) * a + x0
    x.append(x_)
    y1.append((math.sqrt(abs(1 - ((x_-x0) / a)**2) * b**2) + y0))
    y2.append(-((math.sqrt(abs(1 - ((x_-x0) / a)**2) * b**2) - y0)))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
print('Для построения гиперболы введите координаты центра, размеры a и b (x0, y0, a, b):')
x0, y0, a, b = [float(x) for x in input().split(',')]

#x0 = -2
#y0 = -3
x1 = []
x2 = []
y1 = []
y2 = []
#a = 5
#b = 3

fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(8, 8)
plt.axis([-10, (10), -10, 10])
plt.grid()

for i in range(1001):
    x_ = (i / 1000) *  2 * a + x0
    if x_ >= (a + x0):
        x1.append(x_)
        x2.append(-x_)
        y1.append((math.sqrt(abs(1 - ((x_-x0) / a)**2) * b**2) + y0))
        y2.append(-((math.sqrt(abs(1 - ((x_-x0) / a)**2) * b**2) - y0)))
    
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.plot(x2, y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
fig = figure()
ax = Axes3D(fig)
fig.set_size_inches(8, 8)
size_1 = 3
ax.axis([-size_1, size_1, -size_1, size_1])
ax.set_zlim(-size_1, size_1)

a = 3
b = 2
c = 1
X = np.arange(-5, 5, 0.01)
Y = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt((1 - (X/a)**2 - (Y/b)**2) * c**2)

print('Элипс')
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, -Z)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```

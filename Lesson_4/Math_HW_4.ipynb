```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(8, 8)
plt.axis([0, 10, 0, 10])
plt.grid()

A = [3, 4]
b = 4
alfa = 1/6 * np.pi
B = [A[0] + np.cos(alfa) * b , A[1] + np.sin(alfa) * b]
x = [A[0], B[0]]
y = [A[1], B[1]]

print(f'точка А - {A}, угол альфа (в рад) - {alfa}, длина отрезка - {b}, точка В - {B}')

plt.plot(x, y)
plt.plot(6, 2)
plt.xlabel('x')
plt.ylabel('y')
```

### Задание 4

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(8, 8)
plt.axis([0, 0.05, -100, 600])
plt.grid()

a = np.linspace(0.01, 0.02, 101)
k = 1
while True:
    a = np.linspace(0.01, 0.02, 101)
    x = k / a * np.pi
       
    condlist = [(100 < x) & (x < 500)]
    choicelist_x = [x]
    x = np.select(condlist, choicelist_x, None)
    
    if (np.where(x != None)[0].size == 0):
        break
    
    print(f'k = {k}, x_{k}_min = {np.min([i for i in x if i])}, x_{k}_max = {np.max([i for i in x if i])}')
    choicelist_a = [a]
    a = np.select(condlist, choicelist_a, None)
    print(f'k = {k}, a_{k}_min = {np.min([i for i in a if i])}, a_{k}_max = {np.max([i for i in a if i])}')
    
    plt.plot(a, x)
    k += 1
    
plt.ylabel('x')
plt.xlabel('a')
```

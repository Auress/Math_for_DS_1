```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 101)
k1 = 1
k2 = 3
y1 = np.cos(k1 * x)
y2 = np.cos(k2 * x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)

fig.set_size_inches(10, 6)

plt.plot(x, y1, label="cos(x)", color="green")
plt.plot(x, y2, label="cos(3*x)", color="red")

plt.title("Графики функций", fontsize=16)
plt.xlabel("Переменная x", fontsize=14)
plt.ylabel("Функция y(x)", fontsize=14)

plt.axis([0, 10, -2, 2])

legend = plt.legend(loc="lower right", fontsize=14)
plt.setp(legend.get_texts(), color="DarkBlue")

plt.show()
```
